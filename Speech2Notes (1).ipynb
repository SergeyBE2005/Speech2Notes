```python
import os
import time
from google.colab import files
from faster_whisper import WhisperModel

GROQ_API_KEY = "GROQ_API_KEY"
AUDIO_PATH = "/content/drive/MyDrive/Colab Notebooks/Notes-maker/Data/ll_short.m4a"
```

```python
print("Распознаём речь...")
model = WhisperModel("large-v3", device="cuda", compute_type="float16")
segments, _ = model.transcribe(AUDIO_PATH, language="ru", beam_size=5)

transcription = " ".join([seg.text for seg in segments])
print(f"Транскрибация готова ({len(transcription)} символов)")

with open("transcription.txt", "w", encoding="utf-8") as f:
    f.write(transcription)
```

```python
from groq import Groq

client = Groq(api_key=GROQ_API_KEY)

def create_summary(text):
    print("Составляем конспект через Groq...")

    prompt = f"""
    Ты — ассистент профессора. Составь КРАТКИЙ конспект лекции на русском языке.

    Требования:
    - Разбей на смысловые блоки с подзаголовками
    - Выдели **жирным** ключевые термины и определения
    - Используй маркированные списки для важных тезисов
    - В конце напиши раздел "Главные выводы" (3-5 пунктов)
    - Удали слова-паразиты, повторы и "воду"

    Текст лекции:
    {text[:15000]}  # Ограничение по токенам Groq
    """

    chat_completion = client.chat.completions.create(
        messages=[{"role": "user", "content": prompt}],
        model="llama-3.1-8b-instant",
        temperature=0.3
    )

    return chat_completion.choices[0].message.content

summary = create_summary(transcription)
print("\n" + "="*60)
print("КОНСПЕКТ ЛЕКЦИИ:")
print("="*60)
print(summary)

with open("summary.txt", "w", encoding="utf-8") as f:
    f.write(summary)
```
